# Pandas Tutorial
pandas is a column-oriented data analysis API. It's a great tool for analyzing input data that is used by many machine learning frameworks.<br> In this quick workshop I will teach you the basic functions of pandas in order to be able to better manipulate your data

## Importing pandas
- Whenever we want to import pandas we usually write it at the top of our code as seen below
- we give it the alias pd for convention and anytime we wish to call it we refer to it as pd

```python
import pandas as pd
pd.__version__
```

## Structure
Pandas utilizes two main data structures
- **Series:** A series is a single column
- **DataFrame:** A dataframe is a datable made up of one or more series

Let's start by creating a series object:

```python
cities = pd.Series(['San Diego', 'Los Angeles', 'Sacramento', 'San Francisco']) #Create a new series
cities #print series
```

The DataFrame object is can be created by passing a dictionary that maps string column names to their respective series.<br> If the series don't match in length, NA/NaN values will fil in the missing spots. Lets create a dataframe:

```python
population = pd.Series([234234, 100232, 50000]) #create a new series

df = pd.DataFrame({'City': cities, 'City Population':population}) #create dataframe from series
df #print dataframe
```

## Column Manipulation
- There are different ways to access and modify different columns
- Pandas offers a myriad of ways to manipulate its data but I will show you the most common

### Let's start off by accessing a single column

```python
df['Headquarters'] #Here we supply dataframe with the name of the column to access that single column
```

### Above we used bracket notation
- Bracket notation is typically used if your column name has Spaces or funky characters like periods
- Below we use dot notation which performs the same task but instead we use a dot

```python
df.Headquarters
```

### Now lets manipulate rows
- in pandas we can use loc functions in order to specify a single rows , a type of row, or all rows
- This concept is usually a little tough to wrap your head around but with practice it becomes much easier

If we wanted to grab the first 5 rows from the headquarters data how would we do that?

```python
df.loc[0:7 , 'Headquarters'] #[Specify rows , Specify Columns]
```
