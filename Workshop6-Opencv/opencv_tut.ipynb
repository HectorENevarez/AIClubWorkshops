# OpenCV Tutorial
- This is a beginner OpenCV tutorial that will help to teach the basics.

Lets start it off by importing the packages we'll be using this tutorial:

```python
import matplotlib.pyplot as plt #Used in order to print out images
import cv2 #this is opencv
import imutils #This package helps make opencv functions more simplified
```

## What is an image
- To you us it is easy to describe an image, but how does a computer see an image
- Computers see images as numbers in a multi-dimensional array with each (X, Y) representing a pixel

![3-D array](ThreeChannelArray.png)

- Opencv Takes this numerical data and allows us to modify it
- Colors images have 3 dimensional arrays while

## Loading images
- it is easy to load images in opencv
- We can assign a variable an image

```python
image = cv2.imread('landscape.jpg') #Load image 
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #opencv stores in bgr rather than rgb so we need to convert
(height, width, depth) = image.shape #Here we can obtian the shape information about the image
print('width={}, height={}, depth={}'.format(width, height, depth)) #Print image information
```

## Printing Images
- In order to print images we can use opencv directly but for this protion we will utilize matplotlib

```python
plt.imshow(image) #Load image into pyplot
plt.show() #Show image
```

## Cropping images
- In order to crop images we must specify the two opposite corners we want crop the image from

```python
#  image[StartY:endY, Startx:EndX]
roi = image[60:160, 320:430]
plt.imshow(roi) #Load roi into pyplot
plt.show() #Print 
```

## Resizing images
- In order to resize an image we specify the pixel ratio we want
- the function resize takes two inputs: resize(image_name, (pixel size))

```python
resized_image = cv2.resize(image, (200,200)) #Resizing to 200 x 200 pixels
plt.imshow(resized_image) #load image
plt.show()#print
```

## Rotating Images
- Using imutils we can easily rotate an image
- If we want to rotate without cliping imutils can help us with that as well

```python
rotated = imutils.rotate(image, -45) #45 degrees clockwise
plt.imshow(rotated)
plt.show()
```

```python
rotated_noclip = imutils.rotate_bound(image, 45) #45 degrees clockwise
plt.imshow(rotated_noclip)
plt.show()
```

## Drawing on images
- Drawing on images in opencv is easy
- Before drawing on an image we want ot make a copy in order to not modify the original image
- Below we will see how to draw a square, circle, line, and add text

```python
image_draw = image.copy() #copy to not mess up original image

#   rectangle(image_name, (pt1),    (pt2),      (color),   Thickness)
cv2.rectangle(image_draw, (320,60), (420, 160), (0,0,255), 3)
plt.imshow(image_draw)
plt.show()
```

```python
image_circle = image.copy()
#   circle(image_name,   (center),(radius), (color), thickness)
cv2.circle(image_circle, (300,150), 70, (255, 0, 0), -1)
plt.imshow(image_circle)
plt.show()
```

```python
image_line = image.copy()
#   line(image_name, (pt1),   (pt2),     (color),   thickness)
cv2.line(image_line, (60,20), (400,200), (0,255,0), 3)
plt.imshow(image_line)
plt.show()
```

```python
image_text = image.copy()
#   putText(image_name,  text,          (Start pt), font                scale,(color)    Thickness)
cv2.putText(image_text, "SDSU AI Club", (10,25), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)
plt.imshow(image_text)
plt.show()
```
