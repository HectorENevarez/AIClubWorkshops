## Column Manipulation
- There are different ways to access and modify different columns
- Pandas offers a myriad of ways to manipulate its data but I will show you the most common

### Let's start off by accessing a single column

```python
df['Headquarters'] #Here we supply dataframe with the name of the column to access that single column
```

### Above we used bracket notation
- Bracket notation is typically used if your column name has no Spaces or funky characters like periods
- Below we use dot notation wihich performs the same task but instead we use a dot

```python
df.Headquarters
```

### Now lets manipulate rows
- in pandas we can use loc functions in order to specify a single rows , a type of row, or all rows
- This concept is usually a little tough to wrap your head around but with practice it becomes much easier

If we wanted to grab the first 5 rows from the headquarters data how would we do that?

```python
df.loc[0:4 , 'Headquarters'] #[Specify rows , Specify Columns]
```
