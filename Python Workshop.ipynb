# Syntax

### Comments and Semi-Colons
Comments are denoted using two different symbols
- "#" denotes a single line comment
- ''' denotes a multi line comments

Python does not require the use of semi-colons at the end of a line of code
- Python is based off indentation
- Blocks of code are represented with indentation

```python
"""
Multi-line comment
"""
number = 100 #notice there is no semicolon at the end

if number == 100:
    number = 5
    print(number)
```

### Variable Assignment
Python Assumes its data type
- Other programming langauges like Java, C, C++ require you to specify a data type while python does not
- A variable can be reassigned to a different data type


```python
variable = 10
print(type(variable))

variable = "Python is so easy"
print(type(variable))
```

### Common Operators

### If Statements
If statements are used to determine if statements are true
- If a statement is true the if statement block will execute
- The if statement uses indentation in order to know what to execute
- for if-else chains, an else if is written as "elif" and an else statement is written as "else"
- Basic supported logical conditions are == , != , < , <= , > , and >= 

```python
number = 9

if number > 10:
    print("Python is so easy")
    number = number - 10

#This elif will only execute if the first if statement is false
elif number < 10:
    print("This is an if else statement")
    number = number + 10
    
#This else will only execute if all previous if statements are false
else:
    print("This is an else statment")
    number = 0
```
